```python
interval=[]
for j in range(len(userlist)):
    singleuserdate=df_short.loc[df_short['user']==userlist.iloc[j]];
    i=1
    while(i<len(singleuserdate)):
        interval.append([singleuserdate['time'].iloc[i-1],singleuserdate['user'].iloc[i-1],singleuserdate['item'].iloc[i-1],datetime.strptime(singleuserdate['time'].iloc[i],'%Y-%m-%d %H:%M:%S')-datetime.strptime(singleuserdate['time'].iloc[i-1],'%Y-%m-%d %H:%M:%S')]);
        i=i+1;
```

```python
##loop for getting the watching time after filtering the inactive action
sumwatchtime=[]
for j in range(len(userlist)):
    singleuserdate=WI.loc[WI['user']==userlist.iloc[j]];
    i=1
    base=0
    accum=timedelta(seconds=0)
    while(i<len(singleuserdate)):
        while(i<len(singleuserdate) and singleuserdate['item'].iloc[i]==singleuserdate['item'].iloc[base]):
            accum+=singleuserdate['interval'].iloc[i];
            i=i+1;
        if(i==len(singleuserdate)):
            break
        sumwatchtime.append([singleuserdate['time'].iloc[base],singleuserdate['user'].iloc[base],singleuserdate['item'].iloc[base],singleuserdate['interval'].iloc[base]+accum]);
        accum=timedelta(seconds=0)
        base=i;
        i=i+1;
```

```python
summary=pd.DataFrame(sumwatchtime,columns=['time','user','item','timelength'])
```
